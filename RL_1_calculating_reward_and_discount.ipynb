```python
# Step 1: Import required libraries
import gymnasium
import numpy as np
```

```python
# Step 2: Create the environment
env = gymnasium.make("CartPole-v1")
```

```python
# Step 3: Reset environment to start new episode
state = env.reset()[0]
```

```python
# Step 4: Initialize variables
done = False                # To check if episode finished
rewards = []                # To store rewards at each step
```

```python
# Step 5: Run one episode
while not done:

    # Step 5.1: Select random action
    action = env.action_space.sample()

    # Step 5.2: Perform action
    next_state, reward, done, _, _ = env.step(action)

    # Step 5.3: Store reward
    rewards.append(reward)

    # Step 5.4: Move to next state
    state = next_state
```

```python

# Step 6: Close environment
env.close()
```

```python
# Step 7: Calculate Total Reward
total_reward = sum(rewards)
print("Total Reward:", total_reward)
```

```python
# Step 8: Calculate Discounted Reward
def calculate_discounted_reward(rewards, gamma=0.99):
    G = 0
    for r in reversed(rewards):
        G = r + gamma * G
    return G

discounted_reward = calculate_discounted_reward(rewards)

print("Discounted Reward:", discounted_reward)
```
